```python
import pandas as pd

df = pd.read_excel('KPITT_28Dez.xlsx')
df
```

```python
# Converter minutos totais para formato HH:MM:SS

# Verificar se a coluna 'class duration' já está em formato de horas
# Supondo que 60 minutos ou menos sejam considerados horas

df['Class Duration'] = df.apply(lambda row: pd.to_datetime(row['Class Duration'], unit='m').strftime('%H:%M:%S') 
                                if pd.to_numeric(row['Class Duration'], errors='coerce') > 60 else row['Class Duration'], axis=1)
df
```

```python

df.to_excel('KPITT_28Dez_modificado.xlsx', index=False)
```
